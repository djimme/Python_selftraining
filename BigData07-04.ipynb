### 적합성 검정

```python
import pandas as pd

df_titanic = pd.read_csv('./data/titanic.csv')
df_titanic.info()
```

```python
df_titanic
```

```python
df_t = df_titanic[df_titanic['survived']==1]
df_t
```

```python
table = df_t['sex'].value_counts()
table
```

```python
from scipy.stats  import chisquare

chi = chisquare(table, f_exp=[171,171])
print('<===>\n', chi)
 
```

### 독립성 검정

```python
import pandas as pd
```

```python
df_titanic = pd.read_csv('./data/titanic.csv')
df_titanic
```

```python
df_table = pd.crosstab(df_titanic['class'], df_titanic['survived'])
df_table
```

```python
from scipy.stats import chi2_contingency
chi, p, df, expect = chi2_contingency(df_table)
print('Statistic: ', chi)
print('p-value: ', p)
print('df: ', df)
print('expect: \n', expect)
```
