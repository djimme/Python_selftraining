## 3장 4절 산점도(Scatter Plot)

### 2. 산점도 그리기

```python
import pandas as pd 
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
df_iris = pd.DataFrame(data = iris.data, columns=iris.feature_names)
df_iris
```

```python
df_iris['class'] = iris.target
df_iris['class'] = df_iris['class'].map({0:'Setosa',1:'Versicolour',2:'Virginica'})
df_iris
```

```python
plt.title('iris scatter')
plt.xlabel('sepal length(cm)')
plt.ylabel('sepal width(cm)')

plt.scatter(x=df_iris['sepal length (cm)'], y=df_iris['sepal width (cm)'], alpha=0.5)
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=df_iris, hue='class', style='class')
plt.show()
```
