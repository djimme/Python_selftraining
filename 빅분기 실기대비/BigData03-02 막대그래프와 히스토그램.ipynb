## 3장 2절 막대그래프와 히스토그램

### 1. 막대그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
wine_load = load_wine()
```

```python
df_wine = pd.DataFrame(data = wine_load.data, columns = wine_load.feature_names)
df_wine
```

```python
df_wine['class'] = wine_load.target
df_wine['class'] = df_wine['class'].map({0:'class_0', 1:'class_1', 2:'class_2'})
df_wine
```

```python
wine_type = df_wine['class'].value_counts()
wine_type
```

```python
plt.bar(wine_type.index, wine_type.values, width=0.8, bottom=None, align='center')
plt.show()
```

```python
type(wine_type)
```

```python
plt.barh(wine_type.index, wine_type.values, height=0.8, left=None, align='edge')
plt.show()
```

### 2. 히스토그램

```python
plt.title('Wine alcohol histogram')
plt.hist('alcohol', bins=8, range=(11,15), color='purple',data=df_wine)
plt.show()
```
