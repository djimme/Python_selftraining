```python
score = pd.DataFrame(data={'국어':[100,80],'수학':[75,90],'영어':[90,95]}, index=['장화','홍련'])
score
```

```python
score.loc['홍련','영어']=100
score['국어'] = score['국어']-5
score
```

```python
new_students = pd.DataFrame(data={'국어':[70,85], '수학':[65,100],'영어':[95,65]}, index=['콩쥐', '팥쥐'])
score = score.append(new_students)
score
```

```python
new_students1 = pd.Series(data={'국어':85, '수학':55, '영어':80}, name = '해님')
score = score.append(new_students1)
score
```

```python
science = [80,70,90,85,75]
score['과학']=science
score['학년']=1
score
```

```python
score['과학'] = score['과학']+5
score
```

```python
score['총점'] = score['국어']+score['수학']+score['영어']+score['과학']
score
```
