## 3장 6절 상관관계 시각화

### 1. 산점도 행렬

```python
import pandas as pd 
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['class'] = iris.target
df_iris['class'] = df_iris['class'].map({0:'setosa',1:'versicolour',2:'virginica'})
df_iris
```

```python
scatter_matrix(df_iris, alpha=0.5, figsize=(8,8), diagonal='hist')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(df_iris, diag_kind='auto', hue='class')
plt.show()
```

### 2. 상관계수 행렬 그래프

```python
df_iris_corr = df_iris.drop(columns='class').corr(method='pearson')
sns.heatmap(df_iris_corr, xticklabels=df_iris_corr.columns, yticklabels=df_iris_corr.columns, cmap='RdBu_r', annot=True)
plt.show()
```
